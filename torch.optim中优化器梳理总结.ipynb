## 二、Adagrad

1.torch.optim.Adagrad(params, lr=0.01, lr_decay=0, weight_decay=0, initial_accumulator=0, eps=1e-10)

2.能够自适应的调整学习率，在梯度大的地方降低学习率，在梯度小的地方放大学习率
3.缺点：学习率是单调递减的，而且是累加，会导致学习率提前为0，导致网络训练不充分；需要设置一个全局的初始学习率

## 三、Adadelta

1.torch.optim.Adadelta(params, lr=1.0, rho=0.9, eps=1e-6, weight_decay=0)

2.Adadelta:对adagrad进行改进，一是梯度平方采用指数平均，舍弃历史久远值， 二是采用Hessian矩阵的对角线近似Hessian矩阵

## 四、RMSPROP

1.torch.optim.RMSprop(params, lr=0.01, alpha=0.99, eps=1e-8, weight_decay=0, momentum=0,centered=False)

2.RMSprop将Adagrad中梯度累加变为指数平均，可以人为抛弃历史久远的梯度，避免学习率衰减严重
具体做法为：梯度平方求和采用指数加权移动平均，再采用momentum进行梯度矫正
