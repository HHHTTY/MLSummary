```python
# relu激活函数图像
from torch.nn import functional as F
input = torch.arange(-10, 10, 0.01)
output = F.relu(input)
plt.plot(input, output)
plt.title('relu激活函数')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

### 4.relu改进版本Leaky ReLU  
1.Leaky ReLU又称PReLU，是对ReLU激活函数的改进版本
2.函数公式：  
3.函数导数：  
4.Leaky ReLU改变了ReLU激活函数在负输入梯度消失的问题

```python
# leaky relu函数
input = np.arange(-10, 10, 0.01)
def leaky_relu(x, a):
    return max(0, x)+a*min(0, x)
output = []
for x in input:
    output.append(leaky_relu(x, 0.1))
plt.plot(input, output)
plt.grid(True)
plt.show()
```
