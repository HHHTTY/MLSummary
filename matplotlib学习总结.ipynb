#### Matplotlib.pyplot

```python
# 导入库
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# plt.与ax.的区别
# plt
fig=plt.figure(num=1, figsize=(3,3))
plt.plot(range(3), range(3))
plt.show()
```

```python
# ax.
fig=plt.figure(num=1, figsize=(3,3))
ax = fig.add_subplot(111)
ax.plot(range(2), range(2))
plt.show()
```

```python
## rcParam设置测试
import matplotlib as mpl
fig = plt.figure(num='rctest', figsize=(6, 4), dpi=100)
mpl.rcParams['font.family'] = 'Kaiti'
mpl.rc('font', style='italic', size=10)
ax1 = fig.add_subplot(121)
ax1.plot([1, 3, 4], [2, 5, 6])
# 设置x,y轴范围
ax1.set_xlim(1, 7, 1)
ax1.set_ylim(1, 10)
# 设置x,y显示刻度
ax1.set_xticks(np.linspace(1, 10, 10))
ax1.set_xticklabels(['星期一'])
ax1.set_title('测试')
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot([10, 11, 12], [3, 4, 5])
ax2.set_title('验证')
plt.show()
```
